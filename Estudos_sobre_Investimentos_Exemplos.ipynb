**Exemplo 1 - Variação no Poder de Compra**

Supondo que a inflação em determinado período foi de 50%, qual foi a alteração do poder de compra da moeda nesse mesmo intervalo de tempo?

em T1: RS 100,00 / RS 1,00 => compro 100 unidades

em T2: RS 100,00 / RS 1,50 => compro 66,66 unidades

```
def poderdecompra():

  valor_total = 100
  inflacao_periodo = float(input("Digite a inflação acumulada no período:"))

  compra = valor_total/(1*(inflacao_periodo/100)+1)
  poder_de_compra = ((compra/100) - 1) *100

  print(f'Seu poder de compra é de: {round(poder_de_compra,2)}%\nOu seja, se antes você levava 100 produtos para casa, agora você leva {round(compra,2)}')
```

```
poderdecompra()
```

**Exemplo 2.1 - Preço Unitário (PU)**

Pretendo comprar um título que vence em três meses, por um valor que não decidi ainda, que paga juros de 3% do seu valor de face por mês e devolve o principal ao final. Se quero obter uma rentabilidade de 4% ao mês, quanto devo pagar por ele?

Optando pela base 1, tenho:

image.png

```
def PU():

    juros = float(input("Digite os juros pagos pelo título a.m:"))/100
    taxa = float(input("Digite a rentabilidade desejada:"))/100
    meses = int(input("Digite os número de meses para o vencimento do título:"))

    pmt = 1*juros
    pu = 0

    for i in range (1, meses):
      pu += pmt/(1 + taxa)**i

    preco_unitario = pu + ((1+pmt)/(1 + taxa)**meses)
    
    return print(f'O valor a ser pago no título será de: {round(preco_unitario,4)}')
```

```
PU()
```

**Exemplo 2.2 - Preço Unitário (PU)**

Um investidor comprou um LTN (Letra do Tesouro Nacional = Tesouro prefixado) a 8.10 a.a, com um vencimento de 420 dias. Qual é o Preço Unitário desse titulo?

Usando **1000 reais** como valor a receber, temos:

```
def PU2():

    receber = float(input("Digite o valor que deseja receber com o título:"))
    dias_uteis = 252

    meses = int(input("Digite os número de meses para o vencimento do título:"))/dias_uteis
    taxa = float(input("Digite a rentabilidade prefixada a.a:"))/100
                 
    pu2 = receber/(1+taxa)**meses
    
    return print(f'O valor pago no título será de: {round(pu2,4)}')
```
